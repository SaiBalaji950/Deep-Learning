```python
Torchvision

Torchvision is a popular Python library that provides tools and datasets for computer vision tasks. 
It's built on top of PyTorch and offers pre-trained models, common transformations, and utilities for image processing. 
It includes several important modules:

Datasets: Predefined datasets such as ImageNet, CIFAR, COCO, etc., which can be easily loaded and used for training and testing machine learning models.
Models: Pre-trained models such as ResNet, VGG, MobileNet, and others for image classification, 
object detection, and segmentation tasks. These models can be fine-tuned or used directly.

Transforms: Utilities for common image transformations like resizing, cropping, normalization, and augmentation to prepare datasets for training.

Utils: Tools for managing image loading, creating data loaders, and visualization functions.

To install torchvision, you can use:
pip install torchvision
```

```python
Custom Dataset

If you have your custom dataset, 
you can create your own dataset class by subclassing torch.utils.data.Dataset. The __getitem__ and __len__ methods need to be implemented.
```

```python
from torch.utils.data import Dataset
from PIL import Image
class CustomDataset(Dataset):
    def __init__(self,file_paths,labels,transform=None):
        self.file_paths=file_paths
        self.labels=labels
        self.transform=transform
    def __len__(self):
        return len(self.file_paths)
    def __getitem__(self,idx):
        image=Image.open(self.file_paths[idx])
        label=self.labels[idx]
        if self.transform:
            image=self.transform(image)
        return image,label
file_paths=["C:/Users/PythonTeam/Downloads/download.jpg","C:/Users/PythonTeam/Downloads/rc390_3pm.jpg"]
labels=[0,1]
dataset=CustomDataset(file_paths,labels,transform=transforms.ToTensor())
for image,label in dataset:
    print(f'Label: {label}')
    print(f'Image: {image}')
```

```python
Fine-Tuning Pre-trained Models

Fine-tuning a pre-trained model is a common approach when working with a small dataset. 
Instead of training a model from scratch, you can use a pre-trained model and adjust it to fit your specific task.

Fine-Tuning Steps:
Load a pre-trained model.
Replace the final layer to match the number of classes in your dataset.
Freeze earlier layers to prevent their weights from being updated during training, preserving learned features.
Train the model on your dataset.
```

```python
import torch
import torch.nn as nn
import torchvision.models as models
from torch.optim import Adam
model=models.resnet18(pretrained=True)
for param in model.parameters():
    param.requires_grad=False
num_features=model.fc.in_features
model.fc=nn.Linear(num_features,10)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
criterion=nn.CrossEntropyLoss()
optimizer=Adam(model.fc.parameters(),lr=0.001)
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)  
        total += labels.size(0)  
        correct += (predicted == labels).sum().item()  
    accuracy = 100 * correct / total
    print(f"Epoch {epoch + 1}, Loss: {running_loss / len(train_loader):.4f}, Accuracy: {accuracy:.2f}%")
```
